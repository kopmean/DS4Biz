```python
# ฟังก์ชันนับจำนวนคำ

def count_token(list_token):
  words = {}
  for token in list_token:
      if token in words:
          words[token] = words[token] + 1
      else:
          words[token] = 1
  return words
```

```python
# เก็บคำและจำนวนใน list

all_token = []
for t in range(len(df)):
  all_token.extend(df['Tokenized'][t])
counted_token = count_token(all_token)
```

#Tokenize

```python
from pythainlp.util import thai_digit_to_arabic_digit
from pythainlp.ulmfit import lowercase_all
from pythainlp.corpus.common import thai_words
from pythainlp.util import dict_trie
from pythainlp.tokenize import word_tokenize
from pythainlp.util import normalize
from pythainlp.corpus import thai_stopwords
```

```python
# ฟังก์ชันเอาอักษรพิเศษออก
def del_unwant(text):
  return "".join(re.split("^[ก-๏0-๙a-zA-Z0-9\.\\n-]+", text))
```

```python
df['issues'] = df['issues'].apply(lambda x:del_unwant(x))
df['issues']
```

```python
df['issues'] = df['issues'].str.lstrip(' ')
df['issues'] = df['issues'].str.lstrip('123.')
df['issues']
```

```python
df['issues'][1497]
```

```python
# เปลี่ยนเลขไทยเป็นเลขอารบิก
df['text_no_thainum'] = df['issues'].apply(lambda x: thai_digit_to_arabic_digit(x))
df['text_no_thainum'][1497]
```

```python
df['issues'][2209]
```

```python
# เปลี่ยนตัวอักษรอังกฤษให้เป็นพิมเล็ก

df['text_no_thainum'] = lowercase_all(df['text_no_thainum'])
df['text_no_thainum'][2209]
```

```python
# สร้าง thai words dict

custom_dict_del_n = set(thai_words())
trie = dict_trie(dict_source=custom_dict_del_n)
```

```python
# tokenize

df['Tokenized']=df['text_no_thainum'].apply(lambda x :word_tokenize(x,
                                                        engine="attacut",
                                                        custom_dict = trie,
                                                        keep_whitespace=False))
```

```python
# normalize token เช่น การลบวรรณยุกต์ซ้ำ การพิมตัวสะกดหลายๆตัว
def normalize_word(list_word):
  tokens = list_word
  for i in range(len(tokens)):
   tokens[i]=normalize(tokens[i].lstrip())
  return tokens
```

```python
df['normalized_tokenize'] = df['Tokenized'].apply(lambda x:normalize_word(x))
df['normalized_tokenize']
```

```python
# ฟังก์ชันลบคำที่เป็น stopword

def remove_stopwords(list_words):
  tokens=list_words

   # สร้าง list ของ stopword
  stopwords = list(thai_stopwords())
  stopwords.extend(special_char) #เพิ่มอักษรพิเศษที่ได้สำรวจดูตอนแรก
  stopwords.extend([' ','ๆ','',':'])

  for word in tokens:
    if (word in stopwords) or (word.isnumeric()) or (word == ''):
      tokens.remove(word)
  return tokens
```

```python
#ลบคำที่เป็น stopword
for r in range(2):
  df['no_stopwords'] = df['normalized_tokenize'].apply(lambda x:remove_stopwords(x))
```

#TF-IDF

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# ฟังก์ชันว่างเพื่อใส่ใน TF-IDF ใช้เพื่อไม่ต้องการให้ TfidfVectorizer ทำอะไรกับ token แล้ว
def empty_func(doc):
    return doc
```

```python
# สร้าง TF-IDF
tfidf = TfidfVectorizer(analyzer='word',
                        tokenizer=empty_func,
                        preprocessor=empty_func,
                        token_pattern=None,
                        min_df = 0.02
                        )

tfidf_vector = tfidf.fit_transform(df['no_stopwords']) # fit กับข้อความที่ tokenize แล้ว
tfidf_array = np.array(tfidf_vector.todense())
```

```python
# แสดง TF-IDF
df_tfidf = pd.DataFrame(tfidf_array,columns=tfidf.get_feature_names())
df_tfidf
```

#Split Data

```python
# แบ่งข้อมูล X กับ y
X = df_tfidf
y = df.loc[:,'การเดินทาง':'แสงสว่าง']
```

```python
# แบ่ง train test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

#Train model

นำ model ที่รองรับการทำ multilabel มาลองวัดผลเบื้องต้นโดยยังไม่มีการปรับ parameter

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import RadiusNeighborsClassifier

radius_neigh = RadiusNeighborsClassifier()
mlp = MLPClassifier(random_state=0)
rf = RandomForestClassifier(random_state=0)
tree = DecisionTreeClassifier(random_state=0)
knn = KNeighborsClassifier()
```

นำ model ที่สร้างไว้ทำ cross validation เพื่อดูประสิทธิภาพ

```python
from sklearn.model_selection import cross_val_score

models = [radius_neigh,mlp,rf,tree,knn]
classifier = []
mean_accu = []
for m in models:
  scores = cross_val_score(m, X_train, y_train, cv=5)
  classifier.append(m.__class__.__name__)
  mean_accu.append(scores.mean())
```

```python
score_df = pd.DataFrame({'classifier':classifier,
                        'before_tuned_accuracy':mean_accu})
score_df = score_df.sort_values(by='before_tuned_accuracy', ascending=False)
score_df
```

หลังจากวัดประสิทธิภาพแล้วนำโมเดลที่มีประสิทธิภาพสูงสุด 3 อันดับแรกไปปรับค่า parameter

#Tune model

ปรับค่า parameter ด้วย GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

##MLPClassifier

```python
parameters = {'learning_rate_init': np.arange(0.001,0.5,0.01),
              }
mlp_tuned = GridSearchCV(mlp, parameters,cv=5)
mlp_tuned.fit(X_train,y_train)
```

```python
print('คะแนนเฉลี่ยสูงสุด: %0.4f'%(mlp_tuned.best_score_))
```

## Random Forest

```python
parameters = {'n_estimators': [500,100],
              'max_depth': np.arange(10,100,20)
              }
rf_tuned = GridSearchCV(rf, parameters,cv=5)
rf_tuned.fit(X_train,y_train)
```

```python
print('คะแนนเฉลี่ยสูงสุด: %0.4f'%(rf_tuned.best_score_))
```

##KNeighborsClassifier

```python
parameters = {
              'n_neighbors': np.arange(5,100,10)
              }
knn_tuned = GridSearchCV(knn, parameters,cv=5)
knn_tuned.fit(X_train,y_train)
```

```python
print('คะแนนเฉลี่ยสูงสุด: %0.4f'%(knn_tuned.best_score_))
```

#ทดสอบโมเดลที่ปรับค่าแล้ว

```python
# ทำนายข้อมูลจาก test set
y_pred_mlp = mlp_tuned.predict(X_test)
y_pred_rf = rf_tuned.predict(X_test)
y_pred_knn = knn_tuned.predict(X_test)
```

```python
target_names = y_test.columns
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_mlp, target_names=target_names))
```

```python
print(classification_report(y_test, y_pred_rf, target_names=target_names))
```

```python
print(classification_report(y_test, y_pred_knn, target_names=target_names))
```

จากผลการทดสอบข้อมูลจาก test set จะทำการพิจารณาจาก weighted avg เพราะเนื่องจาก class ใน target attribute มีความ imbalance และ weighted avg จะถ่วงน้ำหนักของค่า precision, recall ตามจำนวน support ดังนั้นจะเห็นได้ว่า MLPClassifier มีค่า recall แบบ weighted average มากกว่าโมเดลอื่น แม้ว่า precision จะน้อยกว่า แต่นั้นก็อาจจะหมายถึงโมเดลไม่ได้มีการทำนายโน้มไป class ใด class หนึ่งเป็นหลัก ถ้าหยิบ class 'กีดขวาง' มาดูจะเห็นว่าโมเดลอีก 2 โมเดลที่เหลือจะมีค่า precision ที่สูงแต่ recall น้อยซึ่งอาจหมายถึง 2 โมเดลนั้นทำนายลักษณะคล้ายๆดิ่งข้อสอบ ด้วยเหตุผลนี้จึงคิดว่า MLPClassifier ดีที่สุด
